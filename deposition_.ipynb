```python
# voice_to_text_jypeter_vosk.py
import queue
import threading
import json
import PySimpleGUI as sg

from vosk import Model, KaldiRecognizer
import sounddevice as sd
import sys

sg.theme("DarkBlue3")
layout = [
    [sg.Text("Offline Voice → Text (VOSK)", font=("Helvetica", 16))],
    [sg.Multiline(size=(80,20), key="-TRANSCRIPT-", disabled=True)],
    [sg.Text("Status:"), sg.Text("Idle", key="-STATUS-", size=(40,1))],
    [sg.Button("Start", key="-START-"), sg.Button("Stop", key="-STOP-", disabled=True),
     sg.Button("Save Transcript", key="-SAVE-"), sg.Button("Exit")]
]
window = sg.Window("Voice to Text — jypeter (offline)", layout, finalize=True)

q = queue.Queue()
running = threading.Event()
model_path = "model"  # change if needed

try:
    model = Model(model_path)
except Exception as e:
    sg.popup("Failed to load VOSK model. Make sure 'model' folder exists and contains the unzipped model.", str(e))
    sys.exit(1)

sample_rate = 16000
rec = None
transcript_text = ""

def audio_callback(indata, frames, time_, status):
    if status:
        # put status message
        q.put(("__STATUS__", str(status)))
    q.put(bytes(indata))

def recognition_thread():
    global rec, transcript_text
    rec = KaldiRecognizer(model, sample_rate)
    with sd.RawInputStream(samplerate=sample_rate, blocksize = 8000, dtype='int16',
                           channels=1, callback=audio_callback):
        running.set()
        while running.is_set():
            try:
                data = q.get(timeout=0.5)
            except queue.Empty:
                continue
            if isinstance(data, tuple) and data[0] == "__STATUS__":
                window.write_event_value("-VSTATUS-", data[1])
                continue
            if rec.AcceptWaveform(data):
                res = json.loads(rec.Result())
                text = res.get("text", "")
                if text:
                    transcript_text += text + "\n"
                    window.write_event_value("-NEW_TEXT-", text)
            else:
                # partial (can be used if you want)
                part = json.loads(rec.PartialResult()).get("partial","")
                # we won't show partials to keep UI stable

while True:
    event, values = window.read(timeout=200)
    if event in (sg.WINDOW_CLOSED, "Exit"):
        running.clear()
        break
    if event == "-START-":
        window["-START-"].update(disabled=True)
        window["-STOP-"].update(disabled=False)
        window["-STATUS-"].update("Starting offline recognition...")
        t = threading.Thread(target=recognition_thread, daemon=True)
        t.start()
        window["-STATUS-"].update("Listening (offline)...")
    if event == "-STOP-":
        running.clear()
        window["-STATUS-"].update("Stopping...")
        window["-START-"].update(disabled=False)
        window["-STOP-"].update(disabled=True)
        window["-STATUS-"].update("Stopped")
    if event == "-NEW_TEXT-":
        new = values[event]
        transcript_field = window["-TRANSCRIPT-"].get()
        transcript_field += new + "\n"
        window["-TRANSCRIPT-"].update(transcript_field)
    if event == "-VSTATUS-":
        window["-STATUS-"].update("Audio status: " + values[event])
    if event == "-SAVE-":
        if transcript_text.strip() == "":
            sg.popup("Transcript is empty.")
        else:
            fname = sg.popup_get_file("Save transcript as", save_as=True, default_extension=".txt", file_types=(("Text Files","*.txt"),))
            if fname:
                with open(fname, "w", encoding="utf-8") as f:
                    f.write(transcript_text)
                sg.popup("Saved to", fname)

running.clear()
window.close()
```

```python
import speech_recognition as sr
import requests
import json
from pprint import pprint 


def voice_to_text_for_jupyter():
    """Listens to the microphone and converts the speech to text."""
    # Initialize the recognizer
    r = sr.Recognizer()

    # Use the default microphone as the audio source
    with sr.Microphone() as source:
        print("Listening... Say something!")
        
        # Adjust for ambient noise to improve accuracy
        r.adjust_for_ambient_noise(source, duration=1) 
        
        # Listen for the audio data
        try:
            audio = r.listen(source, timeout=5, phrase_time_limit=10)
        except sr.WaitTimeoutError:
            print("No speech detected. Try again.")
            return

    # Convert the audio to text
    try:
        # Using Google Web Speech API for recognition (requires internet connection)
        text = r.recognize_google(audio)
        
        # Display the transcribed text in the notebook
        print(f"\n✅ You said: {text}")
        return text
    
    except sr.UnknownValueError:
        print("\n❌ Google Speech Recognition could not understand audio.")
        
    except sr.RequestError as e:
        print(f"\n⚠️ Could not request results from Google Speech Recognition service; {e}")

# Run the function in your Jupyter cell
transcribed_text = voice_to_text_for_jupyter()




OPENROUTER_API_KEY = "your_openrouter_api_key_here"  # Replace 'id' with your actual API key
YOUR_SITE_URL = "https://example.com"  # Replace with your site URL
YOUR_SITE_NAME = "My Python App"  # Replace with your site name

# Define the text you want the model to generate legal questions from
# This variable needs to be defined since it's used in your request data.


# --- 2. Construct the API Request ---

# The URL for the OpenRouter chat completions endpoint
url = "https://openrouter.ai/api/v1/chat/completions"

# Headers for the request
headers = {
    "Authorization": f"Bearer ",
    "Content-Type": "application/json",
    "HTTP-Referer": YOUR_SITE_URL,  # Optional. Site URL for rankings on openrouter.ai.
    "X-Title": YOUR_SITE_NAME,      # Optional. Site title for rankings on openrouter.ai.
}

# Data payload for the request
data_payload = {
    "model": "x-ai/grok-4-fast:free",
    "messages": [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "Generate Legal Questions From it?"
                },
                {
                    "type": "text",
                    "text": transcribed_text
                }
            ]
        }
    ],
    # You might want to add 'temperature' or 'max_tokens' here for better control
}

# --- 3. Send the Request and Get the Response ---

try:
    # Send the POST request
    response = requests.post(
        url=url,
        headers=headers,
        data=json.dumps(data_payload) # json.dumps() converts the Python dict to a JSON string
    )

    # Check for a successful response (status code 200)
    response.raise_for_status() 
    
    # Get the JSON content of the response
    response_json = response.json()
    
    # --- 4. Process and Display the Response ---
    
    print("✅ API Request Successful!")
    print("\n--- Raw JSON Response ---")
    pprint(response_json) # Use pprint for a nicely formatted output

    # Extract the generated content (the legal questions)
    if response_json and 'choices' in response_json and len(response_json['choices']) > 0:
        generated_content = response_json['choices'][0]['message']['content']
        print("\n--- Grok's Generated Legal Questions ---")
        print(generated_content)
    else:
        print("\n--- Failed to extract generated content. Check the response_json structure. ---")

except requests.exceptions.HTTPError as errh:
    print(f"🚨 HTTP Error occurred: {errh}")
    print(f"Response status code: {response.status_code}")
    print(f"Response body (for details): {response.text}")
except requests.exceptions.ConnectionError as errc:
    print(f"🚨 Error Connecting: {errc}")
except requests.exceptions.Timeout as errt:
    print(f"🚨 Timeout Error: {errt}")
except requests.exceptions.RequestException as err:
    print(f"🚨 An unexpected error occurred: {err}")

# You can now use the 'transcribed_text' variable for other tasks in your notebook
# print(f"Variable content: {transcribed_text}")
```
